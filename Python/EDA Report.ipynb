 **Introduction**

The dataset contains weather information for multiple Indian cities, including temperature, humidity, wind speed, and other features.

The objective is to understand relationships between variables and prepare for weather type prediction.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

 **Upload CSV File**

```python
df=pd.read_csv("/content/weather_classification_data.csv")
df
```

**Data Overview**



```python
df.shape
print(df.info())
```

**Data Quality Check**

```python
print(df.isnull().sum())
df.duplicated().sum()
display(df.head())
```

 **Univariate Analysis**

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Location', y='Temperature', data=df, ci=None, palette='coolwarm')
plt.title("Average Temperature by Location")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Weather Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Temperature', y='Humidity', hue='Weather Type', data=df)
plt.title("Temperature vs Humidity by Weather Type")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Weather Type'].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette("Set2"))
plt.title("Weather Type Distribution (Pie)")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Season', y='Temperature', data=df)
plt.title("Temperature Distribution by Season")
plt.show()
```

```python
plt.figure(figsize=(18,6))
sns.countplot(x='Weather Type', hue='Humidity', data=df, palette='Spectral')
plt.title("Average Humidity per Weather Type")
plt.show()
```
